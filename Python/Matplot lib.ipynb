```python
fig_obj=plt.figure()

axes=fig_obj.add_axes([0,0,1,1])
axes.plot(x,x**2,'c',label='X_squared')
axes.plot(x,x**3,'r',label='X_cubed')
axes.legend(loc=0)
```

# Plot appearance

```python
fig_obj=plt.figure()

axes=fig_obj.add_axes([0,0,1,1])
axes.plot(x,x**2,'o--',label='X_squared',)
axes.plot(x,x**3,'p--',label='X_cubed')
axes.legend(loc=0)
```

# PLOTTING

```python
s = np.random.poisson(5, 10000)
counts,bins,ignored=plt.hist(s,10,density=True)
plt.show()
```

```python
s = np.random.poisson(5, 10000)
counts,bins,ignored=plt.hist(s,10,density=False)
plt.show()
```

```python
from random import sample
data=sample(range(1,1000),100)
plt.hist(data,density=True)
```
